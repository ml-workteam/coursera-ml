```python
# Какое количество мужчин и женщин ехало на корабле?
data['Sex'].value_counts()
```

```python
# Какой части пассажиров удалось выжить? Посчитайте долю выживших пассажиров. 
# Ответ приведите в процентах (число в интервале от 0 до 100, знак процента не нужен), округлив до двух знаков.
survived_data=data['Survived'].value_counts()

not_survived = survived_data[0]
yes_survived = survived_data[1]

total_passengers = not_survived+yes_survived

survived_part = round(100 * yes_survived/total_passengers, 2)
survived_part
```

```python
# Какую долю пассажиры первого класса составляли среди всех пассажиров? 
# Ответ приведите в процентах (число в интервале от 0 до 100, знак процента не нужен), округлив до двух знаков.
class_data=data['Pclass'].value_counts()
first_class_share = class_data[1]/(class_data[1]+class_data[2]+class_data[3])
first_class_share = round(100*first_class_share, 2)
first_class_share
```

```python
# Какого возраста были пассажиры? 
# Посчитайте среднее и медиану возраста пассажиров. В качестве ответа приведите два числа через пробел.
mean_age = round(data['Age'].mean(), 2)
print(mean_age)
median_age = data['Age'].median()
print(median_age)
```
