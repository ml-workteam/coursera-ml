```python
import pandas as pd
import numpy as nm
```

```python
names= ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids',
    'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
names_of_class = ['Class']
names_of_features = names[1:]
data = pd.read_csv('wine.data', header=None, names = names, index_col = False)
```

```python
# разделим признаки и классы
fdata = data.drop('Class',axis=1) #данные фич
cdata = data.Class #данные классов
#используем масштабирование признаков
from sklearn.preprocessing import scale 
fdata = scale(fdata)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
kf = KFold(n_splits=5, random_state=42, shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = []
scores = []
from sklearn.model_selection import cross_val_score
```

```python
# первый вариант - используем cross_val_score.
# она сама создаст разбиения по переданному KFold, обучит модели и проверит на тестовых выборках
for k in range(1,51):
    knn.append(KNeighborsClassifier(n_neighbors=k))
    scores.append(cross_val_score(knn[k-1], fdata, cdata.values, cv = kf))
```

```python
#средние scores в первом варианте
meanscores = []
for i in range(1,51):
    meanscores.append(scores[i-1].mean())
    print("K=" + str(i) + ": " + str(scores[i-1].mean()))
```

```python
#второй вариант. Делаем цикл по K, а в нем - по разбивкам
#сами обучаем и вычисляем score
knn2 = []
k_accuracy = []
for k in range(1,51):
    knn2.append(KNeighborsClassifier(n_neighbors=k))
    accuracy = []
    for train_index, test_index in kf.split(fdata):
        #print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = fdata[train_index], fdata[test_index]
        y_train, y_test = cdata.loc[train_index], cdata.loc[test_index]
        knn2[k-1].fit(X_train, y_train.values.ravel())
        thescores = knn2[k-1].score(X_test, y_test.values.ravel())
        #print(thescores)
        accuracy.append(thescores)
    k_accuracy.append(nm.mean(accuracy))    
#print (k_accuracy)    
```

```python
accdata = pd.DataFrame(meanscores, columns=['Meth.1'])
```

```python
accdata['Meth.2'] = pd.DataFrame(k_accuracy)
```
